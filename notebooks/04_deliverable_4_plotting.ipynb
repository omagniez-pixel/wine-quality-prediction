# Deliverable 4 - Plotting

```python
from pathlib import Path
import pandas as pd
import numpy as np
PROJECT_ROOT = Path.cwd()
if (PROJECT_ROOT / 'wine_quality_project').exists():
    PROJECT_ROOT = PROJECT_ROOT / 'wine_quality_project'
elif PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent
print('PROJECT_ROOT =', PROJECT_ROOT)
import matplotlib.pyplot as plt
wine = pd.read_csv(PROJECT_ROOT / 'data/processed/wine_quality_combined_with_target.csv')
metrics = pd.read_csv(PROJECT_ROOT / 'results/model_metrics.csv')
```

```python
quality_counts = wine['quality'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
plt.bar(quality_counts.index.astype(str), quality_counts.values, color='#4C78A8')
plt.title('Original Data — Distribution of Wine Quality Scores')
plt.xlabel('Quality score')
plt.ylabel('Number of wines')
plt.tight_layout()
plt.savefig(PROJECT_ROOT / 'plots/plot_1_original_quality_distribution.png', dpi=160)
plt.show()
```

```python
order = ['low quality', 'medium quality', 'high quality']
group_counts = wine['quality_group'].value_counts().reindex(order)
plt.figure(figsize=(8, 5))
plt.bar(group_counts.index, group_counts.values, color=['#D55E00', '#4C78A8', '#009E73'])
plt.title('Feature-Engineered Data — Distribution of Quality Groups')
plt.xlabel('Quality group')
plt.ylabel('Number of wines')
plt.tight_layout()
plt.savefig(PROJECT_ROOT / 'plots/plot_2_engineered_quality_group_distribution.png', dpi=160)
plt.show()
```

```python
plot_metrics = metrics.sort_values('f1_macro', ascending=False)
plt.figure(figsize=(8, 5))
plt.bar(plot_metrics['model'], plot_metrics['f1_macro'], color='#4C78A8')
plt.title('Model Performance — Macro F1-score Comparison')
plt.xlabel('Model')
plt.ylabel('Macro F1-score')
plt.ylim(0, 1.0)
for i, value in enumerate(plot_metrics['f1_macro']):
    plt.text(i, value + 0.01, f'{value:.3f}', ha='center')
plt.tight_layout()
plt.savefig(PROJECT_ROOT / 'plots/plot_3_model_performance_macro_f1.png', dpi=160)
plt.show()
```
