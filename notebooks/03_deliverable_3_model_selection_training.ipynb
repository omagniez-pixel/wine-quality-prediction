# Deliverable 3 - Model Selection and Training

```python
from pathlib import Path
import pandas as pd
import numpy as np
PROJECT_ROOT = Path.cwd()
if (PROJECT_ROOT / 'wine_quality_project').exists():
    PROJECT_ROOT = PROJECT_ROOT / 'wine_quality_project'
elif PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent
print('PROJECT_ROOT =', PROJECT_ROOT)

import sys
sys.path.append(str(PROJECT_ROOT / 'src'))
from data import load_dataset_split
from metrics import compute_metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import joblib
X_train, X_test, y_train, y_test = load_dataset_split()
print(X_train.shape, X_test.shape)
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=200, random_state=42),
    'SVM': SVC(random_state=42),
}
model_files = {
    'Logistic Regression': 'logistic_regression_model.joblib',
    'Random Forest': 'random_forest_model.joblib',
    'SVM': 'svm_model.joblib',
}
rows = []
reports = []
for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    rows.append({'model': name, **compute_metrics(y_test, pred)})
    reports.append(f'===== {name} =====\n')
    reports.append(classification_report(y_test, pred, zero_division=0))
    reports.append('\nConfusion matrix (labels: high quality, low quality, medium quality)\n')
    reports.append(str(confusion_matrix(y_test, pred, labels=['high quality','low quality','medium quality'])))
    reports.append('\n\n')
    joblib.dump(model, PROJECT_ROOT / 'models' / model_files[name])
metrics = pd.DataFrame(rows)
metrics.to_csv(PROJECT_ROOT / 'results/model_metrics.csv', index=False)
(PROJECT_ROOT / 'results/classification_reports.txt').write_text(''.join(reports), encoding='utf-8')
metrics
```

The three models are compared with the same split, the same transformed features, and the same metrics.
